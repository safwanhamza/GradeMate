```python
from transformers import AutoTokenizer, AutoModelForCausalLM

# Load the model and tokenizer locally
model_name = "Qwen/Qwen2.5-7B-Instruct"

# Load tokenizer
tokenizer = AutoTokenizer.from_pretrained(model_name, trust_remote_code=True)

# Load model
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype="auto",  # Automatically selects the best precision
    device_map="auto",   # Automatically assigns model to GPU/CPU
    trust_remote_code=True
)

print("Model and tokenizer loaded successfully.")
```

```python
# Define the input prompt
prompt = "Who are you?"

# Tokenize the input
inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

# Generate output
output_ids = model.generate(
    **inputs,
    max_new_tokens=100,  # Limit the number of tokens in the response
    temperature=0.7,     # Control randomness (lower is more deterministic)
    do_sample=True       # Enable sampling for variability
)

# Decode the generated output
response = tokenizer.decode(output_ids[0], skip_special_tokens=True)

print("Generated Text:")
print(response)
```

```python
# Define the context and question
context = """
In 2010, amidst the rapidly evolving tech landscape, Novatek Dynamics emerged as a pioneering force in enterprise software solutions.
Headquartered in Dubai, the company quickly gained international recognition for its innovative cloud infrastructure, which accelerated
digital transformation for Fortune 500 clients. The company initially faced challenges in convincing traditional enterprises to adopt
AI-driven workflows, but Hassan, the visionary CEO, led the efforts to educate clients on AI's potential to streamline their operations.
"""
question = "What challenges did Novatek Dynamics face in 2010?"




# Create a formatted input prompt
prompt = f"Context:\n{context}\n\nQuestion:\n{question}\n\nAnswer:"

# Tokenize and generate output
inputs = tokenizer(prompt, return_tensors="pt").to(model.device)
output_ids = model.generate(
    **inputs,
    max_new_tokens=200,
    temperature=0.7,
    do_sample=True
)

# Decode the generated output
response = tokenizer.decode(output_ids[0], skip_special_tokens=True)


# Extract the part after "Answer:"
answer_start = response.find("Answer:") + len("Answer:")
final_answer = response[answer_start:].strip()

print(final_answer)

```
